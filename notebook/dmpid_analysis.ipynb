```python
sample_x = []
sample_y = []
click_x = []
click_y = []
data_dict = [0 for i in range(100)]
click_dict = [0 for i in range(100)]
with open("common") as fp:
    for line in fp:
        try:
            dmpid, data = line.strip().split("\t")
            hc_ad_show,hc_ad_click,hc_ps_show,hc_ps_click,wl_ad_show,wl_ad_click,wl_ps_show,wl_ps_click=data.split("-")
            x = int(hc_ps_show)
            y = int(wl_ps_show)+int(wl_ad_show)
            sample_x.append(x)
            sample_y.append(y)
            index = int(min(x/5, y/5))
            data_dict[index] += 1.0
            
            c_x = int(hc_ps_click)
            c_y = int(wl_ps_click)+int(wl_ad_click)
            click_x.append(c_x)
            click_y.append(c_y)
            index = int(min(c_x, c_y))
            click_dict[index] += 1.0
        except Exception:
            pass
```

```python
mid = np.array(data_dict[:100])/25488228
rate = np.array([sum(mid[:i+1]) for i in range(len(mid))])
plt.plot(range(5,505,5), 1-rate, color="red", label="show")

mid = np.array(click_dict[:100])/25488228
rate = np.array([sum(mid[:i+1]) for i in range(len(mid))])
plt.plot(range(1,101), 1-rate, color="blue", label="click")
plt.xlim(0,200,100)
plt.legend()
plt.show()
```
