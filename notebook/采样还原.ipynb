```python
import numpy as np
from matplotlib import pyplot as plt
import random
from matplotlib import cm
```

```python
p=np.arange(0,0.1,0.001)[1:]
p=0.01
r=np.arange(0,1,0.001)[1:]
p1=p/(p+(1-p)/r)
q1=(1+0.02)*p1
q=q1/(q1+(1-q1)*r)
b=q/p-1
```

```python
plt.plot(r, b, color="blue", label="bias")
plt.xlim(0,1,1000)
plt.legend()
plt.show()
```
